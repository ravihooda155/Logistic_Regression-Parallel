```python
import numpy as np
import matplotlib.pyplot as mlt
import pylab
```

### For Iris dataset

```python
iterations=[0,100,200,300,400,500,600,700,800,900,1000]
with_class_0=np.array([0.784246,0.556716,0.443689,0.367582,0.313868,0.274296,0.244044,0.220194,0.200911,0.18499,0.171614])
with_class_1=np.array([0.689623,0.638329,0.636423,0.635155,0.634008,0.632939,0.631925,0.630948,0.629999,0.629071,0.628158])
with_class_2=np.array([0.587374,0.463803,0.418439,0.385708,0.360815,0.341219,0.325341,0.312166,0.301014,0.291416,0.283041])
print(with_class_0.shape)
print(with_class_1.shape)
print(with_class_2.shape)
```

```python

mlt.plot(iterations,with_class_0,label="Iris-setosa")
mlt.plot(iterations,with_class_1,label="Iris-versicolor")
mlt.plot(iterations,with_class_2,label="Iris-virginica")
mlt.title('Iterations vs cross entropy loss for Iris dataset',fontsize=13)
mlt.xlabel('No. of iterations', fontsize=13)
mlt.ylabel('Cross entropy loss', fontsize=13)
pylab.legend(loc='upper right')
mlt.figure(figsize=(20,8))
mlt.show()
```
